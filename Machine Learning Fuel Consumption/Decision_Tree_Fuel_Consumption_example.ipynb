# **Machine Learning for fuel consumption prediction: decision tree model**
Is it possible to predict fuel consumption of vehicles using machine learning? It is a very complex problem. Fuel consumption is affected by many variables such as vehicle mass, engine displacement, tires, aerodynamic, technology, etc.

The objectives of this project are: loading a vehicle database, training a decision tree model, evaluating its performance and testing it with a new vehicle that is not featured in the database.

The machine learning chosen was the decision tree model because of its ability to handle multiple input variables and its decision tree visualization (easy to understand how the decision was made).

The vehicle database: https://www.kaggle.com/prassanth/new-cars-price-2019

The car used for testing the model: https://www.fueleconomy.gov/feg/bymodel/2012_Suzuki_Kizashi.shtml

# **Preparing the database**
In order to run the model, non-numeric values must be converted to numeric through the Labelenconder function.  But this function does not work with empty values, so all lines with empty values are deleted. 

```python
#Drop all lines that contains these empty values. LabelEnconder (numeric conversion of string values) does not work with empty values
df['EPA Fuel Economy Est - City (MPG)'].replace('', np.nan, inplace=True)
df.dropna(subset=['EPA Fuel Economy Est - City (MPG)'], inplace=True)

df['Base Curb Weight (lbs)'].replace('', np.nan, inplace=True)
df.dropna(subset=['Base Curb Weight (lbs)'], inplace=True)

df['Fuel Tank Capacity, Approx (gal)'].replace('', np.nan, inplace=True)
df.dropna(subset=['Fuel Tank Capacity, Approx (gal)'], inplace=True)

df['SAE Net Torque @ RPM'].replace('', np.nan, inplace=True)
df.dropna(subset=['SAE Net Torque @ RPM'], inplace=True)

df['SAE Net Horsepower @ RPM'].replace('', np.nan, inplace=True)
df.dropna(subset=['SAE Net Horsepower @ RPM'], inplace=True)

df['Displacement'].replace('', np.nan, inplace=True)
df.dropna(subset=['Displacement'], inplace=True)

df['Turning Diameter - Curb to Curb (ft)'].replace('', np.nan, inplace=True)
df.dropna(subset=['Turning Diameter - Curb to Curb (ft)'], inplace=True)

df['Stabilizer Bar Diameter - Front (in)'].replace('', np.nan, inplace=True)
df.dropna(subset=['Stabilizer Bar Diameter - Front (in)'], inplace=True)

df['Fuel System'].replace('', np.nan, inplace=True)
df.dropna(subset=['Fuel System'], inplace=True)
```

```python
#Replacing NaN values with '0'
df = df.replace(r'^\s*$', np.NaN, regex=True)
df.fillna(0, inplace = True)
df['EPA Fuel Economy Est - City (MPG)']=df['EPA Fuel Economy Est - City (MPG)'].astype(int)
df['Displacement']=df['Displacement'].astype(float)
```



---

# **Data Preprocessing**
Initially the variables considered to the model were only those with strong correlation seen in the correlation matrix (base curb weight, fuel tank capacity, SAE Net Torque, SAE Net Horsepower, Displacement and Stabilizer Bar Diameter). 

**But as more models were trained, it became clear that choosing all variables untill 'Category' column resulted in better accuracy scores.**

Although increasing the dimensionality of a dataset can deteriorate machine learning performance, this was not the case. 

```python
#LabelEnconder (numeric conversion of string values)
x = df.iloc[:,2:52].values
y = df.iloc[:, 1].values
y=y.astype(int)
labelencoder = LabelEncoder()
x[:,0] = labelencoder.fit_transform(x[:,0])
x[:,1] = labelencoder.fit_transform(x[:,1])
x[:,11] = labelencoder.fit_transform(x[:,11])
x[:,14] = labelencoder.fit_transform(x[:,14])
x[:,15] = labelencoder.fit_transform(x[:,15])
x[:,16] = labelencoder.fit_transform(x[:,16])
x[:,17] = labelencoder.fit_transform(x[:,17])
x[:,43] = labelencoder.fit_transform(x[:,43])
x[:,47] = labelencoder.fit_transform(x[:,47])
x[:,49] = labelencoder.fit_transform(x[:,49])
```



---
# **Spliting the data**


```python
#Splitting the dataframe into train and test. 30% is test
x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(x, y,
                                                                  test_size = 0.3,
                                                                  random_state = 0)
```



---
# **Creating the model and evaluation**
After the spliting of the data, the model is tested and a confusion matrix is printed to compare the prediction values with the true values.  


```python
#Decision Tree model
modelo1 = DecisionTreeClassifier(criterion = 'entropy')
modelo1.fit(x_treinamento, y_treinamento)
export_graphviz(modelo1, out_file = 'modelo1.dot')
previsoes1 = modelo1.predict(x_teste)
confusao1 = ConfusionMatrix(modelo1, fontsize=20)
visualizer = RadViz(classes=modelo1, size=(1080, 720))
confusao1.fit(x_treinamento, y_treinamento)
confusao1.score(x_teste, y_teste)
confusao1.poof()
```

```python
#Accuracy score
accuracy_score(y_teste, previsoes1)
```

**The decision tree model obtained a 92% accuracy and the correlation matrix shows that the incorrect prediction classifications were close to the true value classifications.**
